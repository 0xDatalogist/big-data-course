```
my_tuples = [('sam',23),('mary',34),('peter',25)]
pair_rdd = sc.parallelize(my_tuples)
pair_rdd.collect()
```

```
my_list = [('sam 23'),('mary 34'),('peter 25')]
rdd = sc.parallelize(my_list)
pair_rdd = rdd.map(lambda x : (x.split()[0], x.split()[1]))
pair_rdd.collect()
```

```
scores_tuples = [('messi',23),('ronaldo',28),('neymar',22),('messi',18),('ronaldo',19)]
pair_rdd = sc.parallelize(scores_tuples)
pair_rdd.reduceByKey(lambda x,y: x + y).collect()
```

```
scores_tuples = [('messi',23),('ronaldo',28),('neymar',22),('messi',18),('ronaldo',19)]
pair_rdd = sc.parallelize(scores_tuples)
reduced_pair_rdd = pair_rdd.reduceByKey(lambda x,y: x + y)
reduced_key_reversed_pair_rdd = reduced_pair_rdd.map(lambda x: (x[1],x[0]))
reduced_key_reversed_pair_rdd.sortByKey(ascending=False).collect()
```

```
scores_tuples = [('messi',23),('ronaldo',28),('neymar',22),('messi',18),('ronaldo',19)]
pair_rdd = sc.parallelize(scores_tuples)
pair_rdd.reduceByKey(lambda x,y: x + y).map(lambda x: (x[1],x[0])).sortByKey(ascending=False).collect()
```

```
scores_tuples = [('messi',23),('ronaldo',28),('neymar',22),('messi',18),('ronaldo',19)]
pair_rdd = sc.parallelize(scores_tuples)
pair_rdd.groupByKey().mapValues(list).collect()
```
